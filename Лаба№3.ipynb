```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
df=pd.read_csv('energy.csv')
df = df.dropna()
#Сортировка данных
sorted_df = df.sort_values(by='ТЭС', ascending=True) #сортировка по Алфавиту столбца ТЭС
sorted_df
```

```python
#Добавление к таблице столбца с новыми данными, полученным путем преобразований имеющихся.
mosh = df['Мощность (ГВт)'].mean()
df1 = df.assign(Отклонение_отсреднего=df['Мощность (ГВт)'] - mosh)
df1
```

```python
#Выбор фрагмента таблицы по условию.
#1 способ
condition = df['Мощность (ГВт)'] > 3
filtered_df = df[condition]
filtered_df
```

```python
#2 способ
filtered1_df = df.loc[df['Мощность (ГВт)'] > 4]
filtered1_df
```

```python
#Выбор фрагмента таблицы по номерам строк и столбцов.
filtered_df = df.iloc[[0, 1, 2], [0, 1]] # Выберите строки с индексами 0, 1 и 2, а также столбцы с индексами 0 и 1
filtered_df
```

```python
#Расчеты статистических показателей по сгруппированным данным.
total_sales_by_category = df.groupby('Владелец')['Мощность (ГВт)'].sum()
print(total_sales_by_category)
```

```python
#Построение гистограммы по данным.
plt.bar(df['ТЭС'], df['Мощность (ГВт)'])
plt.gca().yaxis.set_minor_locator(plt.MultipleLocator(0.5))
plt.xticks(rotation=90)
plt.xlabel('ТЭС')
plt.ylabel('Мощность (ГВт)')
plt.title('Гистограмма мощности ТЭС')

plt.show()
```

```python
#Построение круговой диаграммы по данным.
plt.pie(df['Мощность (ГВт)'], labels=df['ТЭС'])
plt.title('Круговая диаграмма мощности ТЭС')
plt.show()
```

```python
#Построение интерактивного графика по данным.
fig = px.bar(df, x='ТЭС', y='Мощность (ГВт)')
fig.show()
```
